## 1. Comparison With The ER Model

We are going to build an ER model for a network having the same number of nodes and edges as our network of artists, i.e. N=10721 nodes and E=12158 edges. We will define a function that computes the probability p of adding an edge between any of the nodes independently.

```python
def probabilityER(nodes, edges):
    """
        Computes the probability of adding an edge between any of the nodes independently, knowing the number of nodes and the number of edges
    """
    return (2*edges)/(nodes*(nodes-1))
```

```python
N = 11488
E = 12148
p = probabilityER(N, E)
```

```python
ER_graph = nx.fast_gnp_random_graph(N, p)
```

### 1.1. Average Degree Of ER Model

```python
sum_ER = 0
for elem in nx.degree(ER_graph):
    sum_ER += elem[1]

avg_degree_ER = sum_ER/N
avg_degree_ER
```

### 1.2. The Density Of The ER Graph

```python
nx.density(ER_graph)
```

### 1.3. The Average Clustering Coefficient

```python
nx.average_clustering(ER_graph)
```

### 1.4. Number of Triangles

```python
sum_ER_triangles = 0

if nx.triangles(ER_graph) !=0 : 
    sum_ER_triangles +=1

sum_ER_triangles
```

### 1.4. Connected Components

```python
nx.components.number_connected_components(ER_graph)
```

## 2. Comparison With The BA Model

Next, we will build a BA model having, again, the same number of nodes and edges of the artists network. This time, we will create a function that computes the number of links m that connect a newly-added node to the nodes that are already in the network.

```python
def compute_m(nodes, edges):
    """
        Computes the number of links through which a new node will be connected to other nodes in the network, knowing the number of nodes N and the number of edges E 
    """
    return edges//nodes
```

```python
N = 11488
E = 12148
m = compute_m(N, E)
int(m)
```

```python
BA_graph = nx.barabasi_albert_graph(N, m)
```

### 2.1. Average Degree Of The BA Model

```python
sum_BA = 0
for elem in nx.degree(BA_graph):
    sum_BA += elem[1]

avg_degree_BA = sum_BA/N
avg_degree_BA
```

### 2.2. The Density Of The BA Model

```python
nx.density(BA_graph)
```

### 2.3. The Average Clustering Coefficient Of The BA Model

```python
nx.average_clustering(BA_graph)
```

### 2.4. Connected Components For The BA Model

```python
nx.components.number_connected_components(BA_graph)
```

### 2.5. Number of Triangles

```python
sum_BA_triangles = 0

if nx.triangles(BA_graph) !=0 : 
    sum_BA_triangles +=1

sum_BA_triangles
```
